```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load the Dataset
file_path = "StudentsPerformance.csv"  # Update with the actual file path
student_data = pd.read_csv(file_path)

# Step 3: Explore the Dataset
print("Dataset Overview:")
print(student_data.head())  # Display the first few rows of the dataset
print("\nData Types:")
print(student_data.dtypes)  # Display data types of each column
print("\nMissing Values:")
print(student_data.isnull().sum())  # Check for missing values

# Step 4: Descriptive Statistics
print("\nSummary Statistics:")
print(student_data.describe())  # Display summary statistics for numerical variables
print("\nFrequency Counts for Categorical Variables:")
for column in student_data.select_dtypes(include=['object']).columns:
    print(student_data[column].value_counts())

# Step 5: Data Visualization
# Example: Histogram of test scores
plt.figure(figsize=(10, 6))
plt.subplot(1, 3, 1)
sns.histplot(student_data['math score'], bins=10, kde=True, color='blue')
plt.title('Math Score Distribution')
plt.xlabel('Math Score')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
sns.histplot(student_data['reading score'], bins=10, kde=True, color='green')
plt.title('Reading Score Distribution')
plt.xlabel('Reading Score')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
sns.histplot(student_data['writing score'], bins=10, kde=True, color='orange')
plt.title('Writing Score Distribution')
plt.xlabel('Writing Score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Example: Box plot of test scores by gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='math score', data=student_data)
plt.title('Math Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='reading score', data=student_data)
plt.title('Reading Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Reading Score')
plt.show()et

plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='writing score', data=student_data)
plt.title('Writing Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Writing Score')
plt.show()
```

```python
# Step 6: Combine the three score columns into a new column 'test score'
student_data['test score'] = student_data['reading score'] + student_data['writing score'] + student_data['math score']

# Step 7: Verify the new column and explore its distribution
print(student_data.head())  # Display the first few rows to verify the new column
print("\nSummary Statistics for Test Scores:")
print(student_data['test score'].describe())  # Display summary statistics for the new column

# Step 8: Data Visualization for the new 'test score' column
plt.figure(figsize=(8, 6))
sns.histplot(student_data['test score'], bins=10, kde=True, color='purple')
plt.title('Distribution of Test Scores')
plt.xlabel('Test Score')
plt.ylabel('Frequency')
plt.show()

# Example: Box plot of test scores by gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='test score', data=student_data)
plt.title('Test Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Test Score')
plt.show()
```
